Machine Learning Part

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
football_data.head()
```

```python
football_data_bkp = football_data
football_data_bkp.to_csv('E:/Important/Study_Material/CA660/football_data_bkp.csv')
```

```python
football_data = football_data_bkp
```

```python
# Get one hot encoding of columns B
one_hot_league = pd.get_dummies(football_data['league'])
one_hot_team = pd.get_dummies(football_data['team'])
# Drop column B as it is now encoded
football_data = football_data.drop('league',axis = 1)
football_data = football_data.drop('team',axis = 1)
# Join the encoded df
football_data = football_data.join(one_hot_league)
football_data = football_data.join(one_hot_team)
```

```python
football_data.head()
```

```python
train, test = train_test_split(football_data, test_size=0.2)
```

```python
test
```

```python
train_X = train.iloc[:, 4:-1].values
print(train_X)
train_y = train.iloc[:, 3].values
print(train_y)
```

```python
test_X = test.iloc[:, 4:-1].values
test_y = test.iloc[:, 3].values
```

```python
forest_regressor = RandomForestRegressor(n_estimators = 500, random_state = 0)
forest_regressor.fit(train_X, train_y)
```

```python
forest_results = np.round(forest_regressor.predict(test_X))
```

```python
forest_results
```

```python
accuracy_score(test_y, forest_results)
```

```python
test_X
```

```python
plt.plot(test_X, forest_regressor.predict(test_X), 
         color = 'green') 
plt.title('Random Forest Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
confusion_matrix(test_y, forest_results)
```

```python
print(classification_report(test_y, forest_results))
```

```python
football_test_data = football_data
football_test_data = football_test_data[football_test_data.year == 2014]
football_test_data = football_test_data[football_test_data.La_liga == 1]
league = 'La Liga'
```

```python
football_test_data
```

```python
test_X_sample = football_test_data.iloc[:, 4:-1].values
test_y_sample = football_test_data.iloc[:, 3].values
test_y_teams = []
```

```python
football_test_data.all()
```

```python
for attr in football_test_data:
    val_present = 0
    
    for val in football_test_data[attr]:
        if (val != 0):
            val_present = 1

    if(val_present == 0):
        football_test_data = football_test_data.drop([attr], axis=1)
        
print(football_test_data)
```

```python
football_test_data = football_test_data.reset_index(drop=True)
#print(football_test_data)
for attr in football_test_data:

    if(attr not in ['year', 'position', 'matches', 'wins', 'scored', 'xG', league]):
        
        for val in range(len(football_test_data[attr])):
            #print(val)
            #print(football_test_data[attr])
            if(football_test_data[attr][val] == 1):
                test_y_teams.append({'team' : attr, 'position' : val + 1})
        
```

```python
test_y_teams = sorted(test_y_teams, key = lambda i: i['position'])
```

```python
predicted_sample = []

for attr in football_test_data:
    
    if(attr not in ['year', 'position', 'matches', 'wins', 'scored', 'xG', league]):
        
        for val in range(len(football_test_data[attr])):
            
            if(football_test_data[attr][val] == 1):
                predicted_sample.append({'team' : attr, 'pred_wins' : forest_test_results[val]})
```
